```python
import numpy as np
import pandas as pd
import tensorflow 
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
from tensorflow.keras.activations import relu,sigmoid,linear
```

```python
from tensorflow.keras.datasets import mnist

(x_train,y_train),(x_test,y_test) = mnist.load_data()
```

```python
x_train = x_train.reshape(60000,784)
print(x_test.shape)
x_test = x_test.reshape(10000,784)
X_train = x_train/255
x_test = x_test/255

x_train.shape
```

```python
from matplotlib import pyplot

for i in range(5):
  pyplot.subplot(333)
  pyplot.imshow(x_train[i].reshape(28,28),cmap='gray')
  pyplot.show()
```

```python
X_train.shape
```

```python

model = Sequential([
    Dense(units=25,activation='relu'),
    Dense(units=15,activation='relu'),
    Dense(units=10,activation='linear')
])
```

```python
model.compile(
    loss=tensorflow.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
    optimizer=tensorflow.keras.optimizers.Adam(learning_rate=0.001),
)

model.fit(
    x_train,y_train,
    epochs=5
)
```

```python
x_train[1].shape
```

```python
for i in range(100,110):
    prediction = model.predict(x_train[i].reshape(1,784))
    prediction_p = tensorflow.nn.softmax(prediction)
    final = np.argmax(prediction_p)

    pyplot.imshow(x_train[i].reshape(28,28),cmap='gray')
    pyplot.title('prediction = '+ str(final) + ' | ' 'actual_val='+str(y_train[i]))
    pyplot.show()
    
    print()
```
